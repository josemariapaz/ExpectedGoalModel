```python
# La columna en la que se encuentran especificados los eventos de cada partido es "categoria evento"
laLiga['categoria_evento'].unique()
```

```python
# Para el trabajo nos interesan unicamente la totalidad de eventos que sean remates al arco, ya sea que estos acaben en gol o no.
# Viendo la documentación del proveedor de datos, estos eventos corresponden a aquellos que tengan pertenezcan a alguna de las
# siguientes categorias: remate fallido, poste, tiro salvado o gol.
remates = laLiga[(laLiga['categoria_evento']=='remate fallido')|
                (laLiga['categoria_evento']=='poste')|
                (laLiga['categoria_evento']=='tiro salvado')|
                (laLiga['categoria_evento']=='gol')]
```

```python
remates.shape
```

```python
remates.head()
```

```python
remates = remates[['fecha','id_evento','id_periodo','minuto','segundo','equipo','calificador','valor','jugador','x','y','outcome','categoria_evento']]
```

```python
remates.head()
```

```python
# Así como está el dataset, cada evento en particular (identificado por la columna "id_evento") está compuseto por varios registros.
# Para el trabajo queremos que cada evento sea un solo registro.
# Primero vemos que valores del campo "calificador" nos interesan.
remates['calificador'].unique()
```

```python
# Nos quedamos con los registros que tienen estos campos
remates = remates[(remates['calificador']=='auto gol')|
                 (remates['calificador']=='remate asistido')|
                 (remates['calificador']=='cabeza')|
                 (remates['calificador']=='penal')|
                 (remates['calificador']=='remate en jugada regular')|
                 (remates['calificador']=='remate en contraataque')|
                 (remates['calificador']=='remate en jugada a balón parado')|
                 (remates['calificador']=='remate a partir de tiro esquina')|
                 (remates['calificador']=='remate de tiro libre')|
                 (remates['calificador']=='remate en jugada de lateral')]
```

```python
remates.head()
```

```python
# Vemos que cuando un evento está presente el "valor" de su "calificador" es un valor nulo
remates.sort_values(['fecha','equipo','id_evento'])
```

```python
# Para evitar confuciones, vamos a transformar estos valores en True
remates.fillna(True, inplace=True)
```

```python
remates.sort_values(['fecha','equipo','id_evento'])
```

```python
remates = remates.set_index(['fecha','id_evento','id_periodo','minuto','segundo','equipo','jugador','categoria_evento','x','y','outcome','calificador'])['valor'].unstack()
remates = remates.reset_index().rename_axis(None).rename_axis(None, axis=1)
remates.fillna(False, inplace=True)
# Lo que se hace es volver una columna a cada valor unico del campo "calificador".
# Esto evidentemente nos llena de columnas con valores nulos (que nosotros los transformamos en False para diferenciarlos), ya
# que solamente unos cuantos calificadores son usados en cada evento en particula (p. ej. un remate cualquiera no puede ser
# de cabeza y penal al mismo tiempo).
```

```python
remates.sort_values(['fecha','equipo','id_evento'])
```

```python
remates[remates['auto gol']==True]
```

```python
# Deseamos descartar los autogoles (que por definición también son remates al arco).
# Para ello descartamos los registos que tengan True en el campo "auto gol"
remates = remates.drop(remates[(remates['auto gol']==True)].index)
```

```python
remates.head()
```

```python
# Finalmente se tiene un dataset de 9245 remates al arco diferentes.
remates.shape
```

```python
# A continuación creamos 2 features adicionales "distancia" y "ángulo"
```

```python
# "distancia" a partir de coordenadas "x" y "y"
remates['distancia'] = np.sqrt((100-remates['x'])**2 + (50-remates['y'])**2)
```

```python
remates.head()
```

```python
# "angulo" también a partir de coordenadas "x" y "y"
remates['dist_a'] = np.sqrt((100-remates['x'])**2 + (54.8-remates['y'])**2)
remates['dist_b'] = np.sqrt((100-remates['x'])**2 + (45.2-remates['y'])**2)
remates['dist_c'] = 9.6
remates['angulo'] = np.degrees(np.arccos((remates['dist_a']**2 + remates['dist_b']**2 - remates['dist_c']**2) / (2*remates['dist_a']*remates['dist_b'])))
```

```python
remates.head()
```

```python
# Creamos la variable "gol" a partir del campo "categoria_evento" 
remates['gol'] = np.where(remates['categoria_evento']=='gol',True,False)
# "gol" es la variable que queremos predecir
```

```python
remates[remates['gol']!=True]
```

```python
# Finalmente nos quedamos unicamente con las variables que utilizaremos para desarrollar el modelo
df = remates[['distancia','angulo','remate asistido','cabeza','penal','remate a partir de tiro esquina','remate de tiro libre','remate en contraataque','remate en jugada a balón parado','remate en jugada de lateral','remate en jugada regular','gol']]
```
