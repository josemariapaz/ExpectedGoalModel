```python
fecha = []
id_evento = []
minuto = []
segundo = []
categoria_evento = []
equipo = []
jugador = []
x = []
y = []
auto_gol = []
remate_asistido = []
cabeza = []
penal = []
remate_jugada_regular = []
remate_contraataque = []
remate_balon_parado = []
remate_tiro_esquina = []
remate_tiro_libre = []
remate_jugada_lateral = []
### acciones de jugada anterior
evento_accion1 = []
x_accion1 = []
y_accion1 = []
minuto_accion1 = []
segundo_accion1 = []

for i in range(len(remates)):
    if remates.index[i]%2 == 1:
        fecha.append(remates['fecha'][i])
        id_evento.append(remates['id_evento'][i])
        minuto.append(remates['minuto'][i])
        segundo.append(remates['segundo'][i])
        categoria_evento.append(remates['categoria_evento'][i])
        equipo.append(remates['equipo'][i])
        jugador.append(remates['jugador'][i])
        x.append(remates['x'][i])
        y.append(remates['y'][i])
        outcome.append(remates['outcome'][i])
        auto_gol.append(remates['auto_gol'][i])
        remate_asistido.append(remates['remate_asistido'][i])
        cabeza.append(remates['cabeza'][i])
        penal.append(remates['penal'][i])
        remate_jugada_regular.append(remates['remate_jugada_regular'][i])
        remate_contraataque.append(remates['remate_contraataque'][i])
        remate_balon_parado.append(remates['remate_balon_parado'][i])
        remate_tiro_esquina.append(remates['remate_tiro_esquina'][i])
        remate_tiro_libre.append(remates['remate_tiro_libre'][i])
        remate_jugada_lateral.append(remates['remate_jugada_lateral'][i])
    else:
        ### acciones de jugada anterior
        evento_accion1.append(remates['categoria_evento'][i])
        x_accion1.append(remates['x'][i])
        y_accion1.append(remates['y'][i])
        minuto_accion1.append(remates['minuto'][i])
        segundo_accion1.append(remates['segundo'][i])
        
remates2 = pd.DataFrame({'fecha':fecha,'id_evento':id_evento,'minuto':minuto,'segundo':segundo,'categoria_evento':categoria_evento,\
                        'equipo':equipo,'jugador':jugador,'x':x,'y':y,'auto_gol':auto_gol,'remate_asistido':remate_asistido,\
                       'cabeza':cabeza,'penal':penal,'remate_jugada_regular':remate_jugada_regular,'remate_contraataque':remate_contraataque,\
                       'remate_contraataque':remate_contraataque,'remate_balon_parado':remate_balon_parado,'remate_tiro_esquina':remate_tiro_esquina,\
                       'remate_tiro_libre':remate_tiro_libre,'remate_jugada_lateral':remate_jugada_lateral,'evento_accion1':evento_accion1,\
                       'x_accion1':x_accion1,'y_accion1':y_accion1,'minuto_accion1':minuto_accion1,'segundo_accion1':segundo_accion1})
```

```python
# A continuación creamos 3 features adicionales "distancia", "ángulo" y distancia_accion1
```

```python
# Primero estandarizamos las columnas x y y para que tengan valores entre 0 y 1
```

```python
remates2['x_est'] = remates2['x'] / 100
remates2['y_est'] = remates2['y'] / 100
remates2['x_accion1_est'] = remates2['x_accion1'] / 100
remates2['y_accion1_est'] = remates2['y_accion1'] / 100
```

```python
# "distancia" a partir de coordenadas "x" y "y"
remates2['distancia_est'] = np.sqrt((1-remates2['x_est'])**2 + (0.5-remates2['y_est'])**2)
```

```python
# "distancia_accion1" a partir de coordenadas "x_accion1" y "y_accion1"
remates2['distancia_accion1_est'] = np.sqrt((1-remates2['x_accion1_est'])**2 + (0.5-remates2['y_accion1_est'])**2)
```

```python
remates2.head()
```

```python
# caculamos "angulo" a partir de coordenadas "x" y "y"
remates2['dist_a'] = np.sqrt((100-remates2['x'])**2 + (54.8-remates2['y'])**2)
remates2['dist_b'] = np.sqrt((100-remates2['x'])**2 + (45.2-remates2['y'])**2)
remates2['dist_c'] = 9.6
remates2['angulo'] = np.degrees(np.arccos((remates2['dist_a']**2 + remates2['dist_b']**2 - remates2['dist_c']**2) / (2*remates2['dist_a']*remates2['dist_b'])))
```

```python
# caculamos "angulo" estandarizado
remates2['angulo_est'] = remates2['angulo'] / 180 
```

```python
remates2.head()
```

```python
# Creamos la variable diff_segundos
remates2['diff_segundos'] = (remates2['minuto']*60 + remates2['segundo']) - (remates2['minuto_accion1']*60 + remates2['segundo_accion1'])
```

```python
# Creamos la variable "gol" a partir del campo "categoria_evento" 
remates2['gol'] = np.where(remates2['categoria_evento']=='gol',True,False)
# "gol" es la variable que queremos predecir
```
